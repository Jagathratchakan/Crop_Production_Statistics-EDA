# EDA on Crop Production Statistics

**Using Plotly**

```python
#Import need packages for handling and visualization the data 
#Data Handling packages
import numpy as np
import pandas as pd
#Data visualization packages
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# Read Dataset

```python
data =  pd.read_csv("/kaggle/input/crop-production-statistics-india/APY.csv")
data.head()
```

```python
data.info()
```

# DATA CLEANING

```python
# Count number of null values present in the data
data.isnull().sum()
```

```python
#Drop the null values fields in the dataset
new_data = data.dropna()
```

```python
#Number of  crops types in the dataset
print("Number of crops : ",new_data['Crop'].nunique())
print("\n------------------xxxxxxxxxxxxxxxxxx-----------------\n")
print("Crops Type : ",new_data["Crop"].unique())
```

```python
#Number of  season in the dataset
print("Number of crops : ",new_data['Season'].nunique())
print("\n------------------xxxxxxxxxxxxxxxxxx-----------------\n")
print("Crops Type : ",new_data["Season"].unique())
```

# Top 5 States Production in India 

```python
print("Top 5 State in production :\n")
top_state = data.groupby('State').sum()['Production'].nlargest()
print(top_state)
```

# Top 5  Production Season in India 

```python
print("Top 5 Production Season :\n")
top_season = data.groupby('Season').sum()['Production'].nlargest()
print(top_season)
```

# Top 5 states in the annul yielding

```python
print("Top 5 State in yield :\n")
top_state = data.groupby('State').sum()['Yield'].nlargest()
print(top_state)
```

# MAXIMUM YIELDS IN INDIA BASED ON STATES

```python
print("Top 5 Areas in the State :\n")
top_area = new_data.groupby('State').agg({"Yield":"max"})
print(top_area)
```

# Visualization for Kerala (**Top in Production Rate**)

```python
kerala = new_data[new_data['State'] == 'Kerala']
kerala.head()
```

```python
# Number of crop production in Kerala
print("Number of crop production : ",kerala["Crop"].nunique())
```

```python
fig = px.histogram(kerala, x="Crop", y="Production", color="Season")
fig.update_layout(barmode='group')
fig.show()
```

```python
fig = px.histogram(kerala, x="Crop", y="Yield", color="Season")
fig.show()
```

```python
fig = px.histogram(kerala, x="Crop_Year", y="Production", color="Crop")
fig.update_layout(barmode='group')
fig.show()
```

# Visualization for TamilNadu (**Top Second in Production Rate and Top in Yeild rate**)

```python
tamil_nadu = new_data[new_data["State"]=='Tamil Nadu']
```

```python
tamil_nadu.info()
```

```python
tamil_yeild = px.histogram(tamil_nadu, x="Crop", y="Yield", color="Season")
tamil_yeild.show()
```

```python
tamil_yeild = px.histogram(tamil_nadu, x="Crop", y="Production", color="Season")
tamil_yeild.show()
```

```python
tn_top_production = tamil_nadu.groupby('Crop').sum()['Production'].nlargest()
print(tn_top_production)
```

```python
tn_top_production = tamil_nadu.groupby(['Crop'])['Production'].sum().sort_values(ascending=False)
print(tn_top_production)
```

```python
tn_top_production = tamil_nadu.groupby(['Season'])['Yield'].sum().sort_values(ascending=False)
print(tn_top_production)
```

```python
tamil_yeild = px.histogram(tamil_nadu, x="Season", y="Yield")
tamil_yeild.show()
```
